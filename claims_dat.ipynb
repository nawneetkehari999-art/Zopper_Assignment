```python
import pandas as pd

df = pd.read_csv("policy_sales_data.csv")

df.head()
```

```python
claim_data = df.copy()

claim_data["Claim_Amount"] = 100000 * 0.10
claim_data["Claim_Date"] = claim_data["Policy_Start_Date"]
claim_data["Claim_Type"] = 1

claims = claim_data.sample(frac=0.3)

claims = claims[[
"Customer_ID",
"Vehicle_ID",
"Claim_Amount",
"Claim_Date",
"Claim_Type"
]]

claims.to_csv("claims_data.csv", index=False)

claim_data.head
```
